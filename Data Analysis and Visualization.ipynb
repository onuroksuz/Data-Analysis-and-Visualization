```python
sns.lmplot(x = "total_bill", y = "tip", hue = "smoker", col = "time", data = df);
```

```python
#buradda col argimanini devreye soktugumuzda iki farkli grafik geldi, degisiklikler cok daha farkli
#ogunlere gorede egimlerin nasil degistigi gozlemlenebiliyor dolayisi ile biz butun durumlari gozununde bulundurmak zorundayiz 
#makina ogenmeden once bizim ogrenmemiz gerekiyor bu arka taraftaki iliskileri.
#evet egimlerde bi fark var ama bu dikkata alacak kadar degerlimi,tip verenlerin sigara icip icmeme durumuna gore miktari degisiyor ama bu iliski anlamlimi
#aradaki farklilik anlamlimi iste bunlari yapacagiz  ve daha bilimsel bi formatta ilerliyecegim
```
